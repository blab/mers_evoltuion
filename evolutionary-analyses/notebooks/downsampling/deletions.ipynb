```python
import json
from augur.utils import json_to_tree
from Bio import Phylo
mut_and_host_info_by_tip_file = '../../intermediate-results-downsampling/mut_and_host_info_by_tip.json'
#read in the tree
with open(mut_and_host_info_by_tip_file, 'r') as f:
    mut_and_host_info_by_tip = json.load(f)
print(mut_and_host_info_by_tip)
```

```python
focus = ["Orf3", "Orf4a"]
deletion_info = {}
for tip, info in mut_and_host_info_by_tip.items():
    deletion_info[tip] = {}
    deletion_info[tip]['region'] = info['region']
    deletion_info[tip]['host'] = info['host']
    deletion_info[tip]['date'] = info['date']
    genes = info['muts_on_path'].keys()
    for gene in genes:
        if gene not in focus:
            continue
        else:
            deletion_info[tip][gene] = info['muts_on_path'][gene]

for tip, info in deletion_info.items():
    if focus not in list(deletion_info[tip].keys()):
        del deletion_info[tip]
```
