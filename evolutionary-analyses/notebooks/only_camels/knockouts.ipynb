```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
knockouts_df = pd.read_csv('results/ko.tsv ', sep='\t')
knockouts_df.to_csv("results/ko.csv")
```

```python
genes = ['Hel', 'Nsp2', 'Nsp3', 'Nsp5', 'Nsp6', 'Orf5', 'Orf8b', 'ExoN', 'N', 'NendoU', 'Nsp4', 'Nsp8', 'Orf4b', 'S1', 'S2', 'Nsp1', 'Nsp7', 'Orf3', 'RdRpb', '2Omethyltransferase', 'M', 'Orf4a', 'E', 'Nsp9', 'Nsp10']
genes = list(map(str.lower,genes))
print(genes)
```

```python
num_ko = {}
columns = np.arange(11, 102, 10).tolist()

for num in columns:
    ko = knockouts_df.iloc[:, num].tolist()  
    colname = knockouts_df.columns[num]      
    gene = colname.replace('_ko', '')        
    num_ko[gene] = 0                        
    for ans in ko:                       
        if ans == 'Yes':                    
            num_ko[gene] += 1

print(num_ko.keys())
num = []
for gene in num_ko:
    num.append(num_ko[gene])
    
print(num)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=num_ko.keys(), y=num)
plt.ylabel('Number of knockouts')
plt.xticks(rotation=90)
plt.show()
```
