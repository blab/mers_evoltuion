```python
import json

with open('../../../intermediate-results-downsampling/isolating_clade/african_mut_and_host_info_by_tip.json', 'r') as f:
    african_mut_and_host_info_by_tip = json.load(f)

with open('../../../intermediate-results-downsampling/isolating_clade/asian_mut_and_host_info_by_tip.json', 'r') as f:
    asian_mut_and_host_info_by_tip = json.load(f)

print(african_mut_and_host_info_by_tip)
```

```python
def simplified(info_by_tip):
    info_by_tip_simplified = info_by_tip
    for tip, info in info_by_tip.items():
        muts_on_path = info.get('muts_on_path')
        for gene in muts_on_path.keys():
            mutations = muts_on_path[gene]
            for mut1 in mutations:
                count = 0
                for mut2 in mutations:
                    if mut1[1:-1] == mut2[1:-1]:
                        count += 1
                    if count > 1:
                        info_by_tip_simplified[tip]['muts_on_path'][gene].remove(mut2)
    return info_by_tip_simplified

african_info_by_tip_simplified = simplified(african_mut_and_host_info_by_tip)
asian_info_by_tip_simplified = simplified(asian_mut_and_host_info_by_tip)

```

```python
with open('../../../intermediate-results-downsampling/isolating_clade/african_mut_and_host_info_by_tip_streamlined.json', 'w') as outfile:
    json.dump(african_info_by_tip_simplified, outfile, indent=2)

with open('../../../intermediate-results-downsampling/isolating_clade/asian_mut_and_host_info_by_tip_streamlined.json', 'w') as outfile:
    json.dump(asian_info_by_tip_simplified, outfile, indent=2)
```
